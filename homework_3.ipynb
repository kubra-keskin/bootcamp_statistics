### 1.dağılım = np.random.dağılım_adı([],[]) kodunu kullanarak, dağılım_adı yazan yere seçtiğiniz bir dağılımı ve parantez içinde [] şeklinde gösterilen boşlukları yine kendi seçiminiz doğrultusunda doldurun. Daha sonra, a) Verinin histogramını çiziniz b) Q-Q grafiğini çizip yorumlayınız

```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
mu = 6 
sigma = 2
n = 10000
plt.figure(figsize=(8,4), dpi = 100)
dag_1 =  np.random.normal(mu, sigma, n)
plt.hist(dag_1, bins = 50)
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4), dpi = 200)
nsample = 1000
res = stats.probplot(dag_1 ,plot=plt)
plt.title("Q-Q Çizgesi")
plt.xlabel("Teorik Değerler")
plt.ylabel("dag_1")
plt.show()
```

```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
mu = 10 
sigma = 5
n = 10000
plt.figure(figsize=(8,4), dpi = 100)
dag_2 =  np.random.normal(mu, sigma, n)
plt.hist(dag_2, bins = 50)
plt.show()
```

```python
dag_1.sort()
dag_2.sort()

plt.figure(figsize=(8,5), dpi = 100)
plt.plot(dag_1, dag_2,"o") 
plt.ylabel('dag_2') 
plt.xlabel('dag_1')
plt.show()
```

```python
plt.figure(figsize=(10,4), dpi = 200)

plt.title('Örnek Veri Kümesi')
plt.xlabel('1.örnek veri kümesi')
plt.ylabel('2.örnek veri kümesi')
plt.scatter(dag_1, dag_2)
plt.plot([np.min(dag_1), np.max(dag_1)],
         [np.min(dag_2), np.max(dag_2)], c='red')


plt.show()
```
